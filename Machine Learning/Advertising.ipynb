```python
#BAC 1
import pandas as pd
url ='/content/Advertising.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()


correlation = data.corr()
print(correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)

data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize = (9,9), color = 'black')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['TV']),y=data['sales'])
print('Value Intercept: ',model.intercept_)
print('Value tri Coeffient: ',model.coef_)

model.score(X=pd.DataFrame(data['TV']),y=data['sales'])

data.plot(kind = 'scatter', x = 'TV', y = 'sales', figsize = (9,9), color = 'black')
y_pred = model.predict(X=pd.DataFrame(data['TV']))
plt.plot(data['TV'],y_pred,color = 'blue')
```

```python
#BAC 2
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'],data['TV']**2]).T
model.fit(X,y=data['sales'])
print('He so Intercept: ',model.intercept_)
print('He so Coeffient: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X,y=data['sales']))

range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X_range)
data.plot(kind = 'scatter',x ='TV', y='sales',figsize=(9,9),color='black',xlim=(0,100))
plt.plot(range,y_pred,color='blue')
```

```python
#BAC 20
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4,data['TV']**5
                  ,data['TV']**6,data['TV']**7,data['TV']**8,data['TV']**9,data['TV']**10
                  ,data['TV']**11,data['TV']**12,data['TV']**13,data['TV']**14
                  ,data['TV']**15,data['TV']**16,data['TV']**17,data['TV']**18
                  ,data['TV']**19,data['TV']**20]).T
model.fit(X,y=data['sales'])
print('He so Intercept: ',model.intercept_)
print('He so Coeffient: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X,y=data['sales']))

range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9
                      ,range**10,range**11,range**12,range**13,range**14,range**15,range**16,range**17
                      ,range**18,range**19,range**20]).T
y_pred=model.predict(X_range)
data.plot(kind = 'scatter',x ='TV', y='sales',figsize=(9,9),color='black',xlim=(0,200))
plt.plot(range,y_pred,color='blue')
```
