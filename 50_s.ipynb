```python
#BAC 1
import pandas as pd
url ='/content/50_Startups.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()


correlation = data.corr()
print(correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)

data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'black')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])
print('Value Intercept: ',model.intercept_)
print('Value tri Coeffient: ',model.coef_)

model.score(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])

data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'black')
y_pred = model.predict(X=pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'],y_pred,color = 'blue')
```

```python
#BAC 2
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2]).T
model.fit(X,y=data['Profit'])
print('He so Intercept: ',model.intercept_)
print('He so Coeffient: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X,y=data['Profit']))

range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X_range)
data.plot(kind = 'scatter',x ='R&D Spend', y='Profit',figsize=(9,9),color='black',xlim=(0,200000))
plt.plot(range,y_pred,color='blue')
```

```python
#BAC 20
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2,data['R&D Spend']**3,data['R&D Spend']**4,data['R&D Spend']**5
                  ,data['R&D Spend']**6,data['R&D Spend']**7,data['R&D Spend']**8,data['R&D Spend']**9,data['R&D Spend']**10
                  ,data['R&D Spend']**11,data['R&D Spend']**12,data['R&D Spend']**13,data['R&D Spend']**14
                  ,data['R&D Spend']**15,data['R&D Spend']**16,data['R&D Spend']**17,data['R&D Spend']**18
                  ,data['R&D Spend']**19,data['R&D Spend']**20]).T
model.fit(X,y=data['Profit'])
print('He so Intercept: ',model.intercept_)
print('He so Coeffient: ',model.coef_)
print('Do chinh xac cua mo hinh: ',model.score(X,y=data['Profit']))

range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9
                      ,range**10,range**11,range**12,range**13,range**14,range**15,range**16,range**17
                      ,range**18,range**19,range**20]).T
y_pred=model.predict(X_range)
data.plot(kind = 'scatter',x ='R&D Spend', y='Profit',figsize=(9,9),color='black',xlim=(0,200000))
plt.plot(range,y_pred,color='blue')
```
