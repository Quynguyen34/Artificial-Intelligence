```python
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd
url ='/content/crabs.csv'
data = pd.read_csv(url)
data.head()

data.replace(['F','M','B','O'],[0,1,0,1])
# speies: O->1 B->0 | sex: M->1 F-1

x = data.drop(['sex'], axis = 1).replace(['F','M','B','O'],[0,1,0,1])
print(x[:20])
y = data.drop(['speies','index','fontallip','rearwidth','length','width','depth'], axis =1).replace(['F','M','B','O'],[0,1,0,1])
print(y[:20])

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train_std)
print(y)
#have scaler
model = Perceptron(max_iter = 200000, eta0= 0.01 , random_state = 0) #number of practice times, speed learn
model.fit(x_train_std, y_train)
y_pred = model.predict(x_test_std)
print(y_pred[:20])
print(y_test[:20])
print("Do chinh xac cua mo hinh:",round(accuracy_score(y_test, y_pred),2))
```
