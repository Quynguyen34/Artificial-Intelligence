```python
#EXERCISE 1: Medical Diagnosis System Using Fuzzy Logic 

import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control  as ctrl
from skfuzzy import membership


#range for program
fever = ctrl.Antecedent(np.arange(1,11,1),'fever')
headache = ctrl.Antecedent(np.arange(1,11,1),'headache') 
nausea = ctrl.Antecedent(np.arange(1,11,1),'nausea')
vomiting = ctrl.Antecedent(np.arange(1,11,1),'vomiting')
jaundice = ctrl.Antecedent(np.arange(1,11,1),'jaundice')
enlargeliver = ctrl.Antecedent(np.arange(1,11,1),'enlargeliver')
jonitpain = ctrl.Antecedent(np.arange(1,11,1),'jonitpain')
bodyweakness = ctrl.Antecedent(np.arange(1,11,1),'bodyweakness')
dizziness = ctrl.Antecedent(np.arange(1,11,1),'dizziness')
lossofappetite = ctrl.Antecedent(np.arange(1,11,1),'lossofappetite')
mp = ctrl.Antecedent(np.arange(1,11,1),'mp')

conclusion = ctrl.Consequent(np.arange(1,11,1),'conclusion')

#Linguistic Variables

#Fuzzy Value for Fever
fever['mild'] = fuzz.trimf(fever.universe,[1,3,4])
fever['moderate'] = fuzz.trimf(fever.universe,[3,5,6])
fever['severe'] = fuzz.trimf(fever.universe,[5,7,8])
fever['very severe'] = fuzz.trimf(fever.universe,[7,10,10])

#Fuzzy Value for Headache
headache['mild'] = fuzz.trimf(headache.universe,[1,3,4])
headache['moderate'] = fuzz.trimf(headache.universe,[3,5,6])
headache['severe'] = fuzz.trimf(headache.universe,[5,7,8])
headache['very severe'] = fuzz.trimf(headache.universe,[7,10,10])

#Fuzzy Value for Nausea
nausea['mild'] = fuzz.trimf(nausea.universe,[1,3,4])
nausea['moderate'] = fuzz.trimf(nausea.universe,[3,5,6])
nausea['severe'] = fuzz.trimf(nausea.universe,[5,7,8])
nausea['very severe'] = fuzz.trimf(nausea.universe,[7,10,10])

#Fuzzy Value for Vomiting
vomiting['mild'] = fuzz.trimf(vomiting.universe,[1,3,4])
vomiting['moderate'] = fuzz.trimf(vomiting.universe,[3,5,6])
vomiting['severe'] = fuzz.trimf(vomiting.universe,[5,7,8])
vomiting['very severe'] = fuzz.trimf(vomiting.universe,[7,10,10])

#Fuzzy Value for Jaundice
jaundice['mild'] = fuzz.trimf(jaundice.universe,[1,3,4])
jaundice['moderate'] = fuzz.trimf(jaundice.universe,[3,5,6])
jaundice['severe'] = fuzz.trimf(jaundice.universe,[5,7,8])
jaundice['very severe'] = fuzz.trimf(jaundice.universe,[7,10,10])


#Fuzzy Value for Enlarge liver
enlargeliver['mild'] = fuzz.trimf(enlargeliver.universe,[1,3,4])
enlargeliver['moderate'] = fuzz.trimf(enlargeliver.universe,[3,5,6])
enlargeliver['severe'] = fuzz.trimf(enlargeliver.universe,[5,7,8])
enlargeliver['very severe'] = fuzz.trimf(enlargeliver.universe,[7,10,10])

#Fuzzy Value for Jonit pain
jonitpain['mild'] = fuzz.trimf(jonitpain.universe,[1,3,4])
jonitpain['moderate'] = fuzz.trimf(jonitpain.universe,[3,5,6])
jonitpain['severe'] = fuzz.trimf(jonitpain.universe,[5,7,8])
jonitpain['very severe'] = fuzz.trimf(jonitpain.universe,[7,10,10])

#Fuzzy Value for Body weakness
bodyweakness['mild'] = fuzz.trimf(bodyweakness.universe,[1,3,4])
bodyweakness['moderate'] = fuzz.trimf(bodyweakness.universe,[3,5,6])
bodyweakness['severe'] = fuzz.trimf(bodyweakness.universe,[5,7,8])
bodyweakness['very severe'] = fuzz.trimf(bodyweakness.universe,[7,10,10])

#Fuzzy Value for Dizziness
dizziness['mild'] = fuzz.trimf(dizziness.universe,[1,3,4])
dizziness['moderate'] = fuzz.trimf(dizziness.universe,[3,5,6])
dizziness['severe'] = fuzz.trimf(dizziness.universe,[5,7,8])
dizziness['very severe'] = fuzz.trimf(dizziness.universe,[7,10,10])

#Fuzzy Value for Loss of loss of appetite
lossofappetite['mild'] = fuzz.trimf(lossofappetite.universe,[1,3,4])
lossofappetite['moderate'] = fuzz.trimf(lossofappetite.universe,[3,5,6])
lossofappetite['severe'] = fuzz.trimf(lossofappetite.universe,[5,7,8])
lossofappetite['very severe'] = fuzz.trimf(lossofappetite.universe,[7,10,10])

#Fuzzy Value for Loss of Mp
mp['mild'] = fuzz.trimf(mp.universe,[1,3,4])
mp['moderate'] = fuzz.trimf(mp.universe,[3,5,6])
mp['severe'] = fuzz.trimf(mp.universe,[5,7,8])
mp['very severe'] = fuzz.trimf(mp.universe,[7,10,10])

#Fuzzy Value for Loss of Conclusion
conclusion['mild'] = fuzz.trimf(conclusion.universe,[1,3,4])
conclusion['moderate'] = fuzz.trimf(conclusion.universe,[3,5,6])
conclusion['severe'] = fuzz.trimf(conclusion.universe,[5,7,8])
conclusion['very severe'] = fuzz.trimf(conclusion.universe,[7,10,10])

#Fuzzy Rule Base for Malaria
rule1 = ctrl.Rule(fever['mild'] & headache['mild'] & nausea['mild'] & vomiting['mild'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['mild'] & bodyweakness['mild'] & dizziness['severe']
                  & lossofappetite['mild'] & mp['mild'], conclusion['mild'])
rule2 = ctrl.Rule(fever['moderate'] & headache['mild'] & nausea['mild'] & vomiting['mild'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['moderate'] & bodyweakness['moderate'] & dizziness['moderate']
                  & lossofappetite['severe'] & mp['moderate'], conclusion['moderate'])
rule3 = ctrl.Rule(fever['severe'] & headache['moderate'] & nausea['mild'] & vomiting['mild'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['mild'] & bodyweakness['severe'] & dizziness['severe']
                  & lossofappetite['severe'] & mp['moderate'], conclusion['severe'])
rule4 = ctrl.Rule(fever['very severe'] & headache['mild'] & nausea['mild'] & vomiting['mild'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['severe'] & bodyweakness['severe'] & dizziness['mild']
                  & lossofappetite['mild'] & mp['severe'], conclusion['very severe'])
rule5 = ctrl.Rule(fever['moderate'] & headache['mild'] & nausea['mild'] & vomiting['moderate'] & jaundice['mild']
                  & enlargeliver['mild'] & jonitpain['moderate'] & bodyweakness['moderate'] & dizziness['moderate'] 
                  & lossofappetite['severe'] & mp['moderate'], conclusion['moderate'])
rule6 = ctrl.Rule(fever['mild'] & headache['moderate'] & nausea['moderate'] & vomiting['mild'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['mild'] & bodyweakness['mild'] & dizziness['moderate'] 
                  & lossofappetite['mild'] & mp['mild'], conclusion['mild'])
rule7 = ctrl.Rule(fever['mild'] & headache['mild'] & nausea['moderate'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['severe'] & bodyweakness['severe'] & dizziness['moderate']
                  & lossofappetite['moderate'] & mp['moderate'], conclusion['severe'])
rule8 = ctrl.Rule(fever['moderate'] & headache['moderate'] & nausea['moderate'] & vomiting['moderate'] & jaundice['moderate'] 
                  & enlargeliver['moderate'] & jonitpain['moderate'] & bodyweakness['moderate'] & dizziness['moderate']
                  & lossofappetite['moderate'] & mp['moderate'], conclusion['moderate'])
rule9 = ctrl.Rule(fever['moderate'] & headache['mild'] & nausea['moderate'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['severe'] & jonitpain['moderate'] & bodyweakness['moderate'] & dizziness['severe']
                  & lossofappetite['moderate'] & mp['moderate'], conclusion['moderate'])
rule10 = ctrl.Rule(fever['mild'] & headache['mild'] & nausea['moderate'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['moderate'] & jonitpain['mild'] & bodyweakness['mild'] & dizziness['mild']
                  & lossofappetite['mild'] & mp['mild'], conclusion['mild'])
rule11 = ctrl.Rule(fever['severe'] & headache['severe'] & nausea['severe'] & vomiting['severe'] & jaundice['severe'] 
                  & enlargeliver['severe'] & jonitpain['severe'] & bodyweakness['very severe'] & dizziness['severe']
                  & lossofappetite['severe'] & mp['moderate'], conclusion['very severe'])
rule12 = ctrl.Rule(fever['moderate'] & headache['severe'] & nausea['moderate'] & vomiting['severe'] & jaundice['moderate'] 
                  & enlargeliver['severe'] & jonitpain['moderate'] & bodyweakness['severe'] & dizziness['moderate']
                  & lossofappetite['mild'] & mp['moderate'], conclusion['severe'])
rule13 = ctrl.Rule(fever['mild'] & headache['moderate'] & nausea['moderate'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['mild'] & bodyweakness['moderate'] & dizziness['mild']
                  & lossofappetite['moderate'] & mp['moderate'], conclusion['moderate'])
rule14 = ctrl.Rule(fever['severe'] & headache['severe'] & nausea['moderate'] & vomiting['mild'] & jaundice['severe'] 
                  & enlargeliver['mild'] & jonitpain['severe'] & bodyweakness['severe'] & dizziness['moderate']
                  & lossofappetite['moderate'] & mp['severe'], conclusion['severe'])
rule15 = ctrl.Rule(fever['mild'] & headache['mild'] & nausea['mild'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['mild'] & bodyweakness['severe'] & dizziness['mild']
                  & lossofappetite['severe'] & mp['moderate'], conclusion['severe'])
rule16 = ctrl.Rule(fever['very severe'] & headache['moderate'] & nausea['mild'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['moderate'] & jonitpain['mild'] & bodyweakness['very severe'] & dizziness['severe']
                  & lossofappetite['mild'] & mp['moderate'], conclusion['very severe'])
rule17 = ctrl.Rule(fever['mild'] & headache['very severe'] & nausea['moderate'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['moderate'] & jonitpain['mild'] & bodyweakness['moderate'] & dizziness['very severe']
                  & lossofappetite['mild'] & mp['moderate'], conclusion['very severe'])
rule18 = ctrl.Rule(fever['moderate'] & headache['very severe'] & nausea['very severe'] & vomiting['mild'] & jaundice['severe'] 
                  & enlargeliver['severe'] & jonitpain['moderate'] & bodyweakness['severe'] & dizziness['very severe']
                  & lossofappetite['very severe'] & mp['severe'], conclusion['very severe'])
rule19 = ctrl.Rule(fever['moderate'] & headache['moderate'] & nausea['moderate'] & vomiting['moderate'] & jaundice['mild'] 
                  & enlargeliver['mild'] & jonitpain['moderate'] & bodyweakness['moderate'] & dizziness['moderate']
                  & lossofappetite['moderate'] & mp['moderate'], conclusion['moderate'])
rule20 = ctrl.Rule(fever['very severe'] & headache['severe'] & nausea['severe'] & vomiting['severe'] & jaundice['severe'] 
                  & enlargeliver['severe'] & jonitpain['severe'] & bodyweakness['severe'] & dizziness['severe']
                  & lossofappetite['severe'] & mp['moderate'], conclusion['very severe'])

#learning code for Malaria
concluding_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20])
concluding = ctrl.ControlSystemSimulation(concluding_ctrl)


#input
concluding.input['fever'] = 2
concluding.input['headache'] = 2
concluding.input['nausea'] = 2
concluding.input['vomiting'] = 2
concluding.input['jaundice'] = 2
concluding.input['enlargeliver'] = 2
concluding.input['jonitpain'] = 2
concluding.input['bodyweakness'] = 2
concluding.input['dizziness'] = 7
concluding.input['lossofappetite'] = 2
concluding.input['mp'] = 3
concluding.compute()
print('Percent Power:',round(concluding.output['conclusion'],2))
conclusion.view(sim=concluding)

```
